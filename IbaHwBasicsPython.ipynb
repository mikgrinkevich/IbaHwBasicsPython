<b>Second task:</b>

Составить список покупок. За один раз можно добавить один товар. Количество товаров не ограничено. (Вводить до тех пор, пока не будет введена пустая строка). После завершения ввода всех необходимых покупок, вывести окончательный список.

```python
print("To exit enter empty field\nPlease, enter some things to buy")
shopping_list = []
while True:
    toBuy = input()
    if(len(str.strip(toBuy))):
        shopping_list.append(toBuy)
    else:
        print(shopping_list)
        break
```

Удалить из списка продукт, чаще всего встречающийся в полученном списке.

```python
shopping_list_dict = {}
while True:
    toBuy = input()
    if(len(str.strip(toBuy))):
        shopping_list.append(toBuy)
    else:
        for i in shopping_list:
            shopping_list_dict[i] = shopping_list.count(i)
            
        #get the most common value in dictionary and remove it by key
        max_key = max(shopping_list_dict, key=shopping_list_dict.get)
        most_common_value = shopping_list_dict.pop(max_key)
        
        #print shopping list
        for key in shopping_list_dict:
            print(key, ' : ', shopping_list_dict[key])
        print("deleted most frequent value: ", max_key)
        break
```
